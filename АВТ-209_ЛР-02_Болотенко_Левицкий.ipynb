```python
def imshowWithAnnotValue(dataImage):
    """
    Отображает черно-белое изображение с просмотром значений пикселей
        - dataImage: матрица значений пикселей
    """
    # Вывод изображения
    maxVal =  dataImage.max()
    minVal =  dataImage.min()
    rangeVal = maxVal - minVal
    print(minVal, maxVal, rangeVal)

    plt.figure(figsize=(10,10)) # создать полотно для рисования, figsize-размер в дюймах
    shw = plt.imshow(# отобразить массив как изображение
        X=dataImage, # данные
        cmap="gray",  # название цветовой шкалы
        vmin=minVal, vmax=maxVal, # границы цветовой шкалы
        interpolation=None, filternorm=True # отключить сглаживание и размытие
    )
    plt.grid(False)
    bar = plt.colorbar(shw) # отобразить цветовую шкалу

    ax = plt.gca()
    nrow = dataImage.shape[0]
    ncol = dataImage.shape[1]

    # Нанести подписей значений для каждого элемента
    for i in range(nrow):
        for j in range(ncol):
            if dataImage[i, j] > 0.3*(rangeVal+minVal):  myColor = "black"
            else:     myColor = "white"
            #print (i, j, myColor, dataImage[i, j])

            text = ax.text(j, i,
                           f"{dataImage[i, j]:.0f}" if maxVal>1 else f"{dataImage[i, j]:.1f}",
                           ha="center", va="center", color=myColor, fontweight="normal", fontsize=7)
    plt.show()  # отрисовать содержимое полотна
```

# Свёрточная НС.

В свёрточных нейронных сетях используются дополнительные слои для понижения размерности исходной матрицы признаков (разрешения изображения). Понижение размерности проводится повторяется многократно, но выполняется двумя видами операций:
- свёртки (convolution)
- объединения (пулинга, pooling)

Группа слоёв свёртки и объединения носит название блока выделения признаков из изображения, т.к. значительно понижает кол-во пикселей, при этом обеспечивает возможность различать исходные изображения. Набор значений, полученных после «блока извлечения новых признаков», предаётся на группу полносвязных слоёв (MLP), где и происходит определение принадлежности к искомым классам.
Таким образом сверхточная нейронная сеть состоит из основных частей:
- "блока извлечения признаков" из изображения
- "блока классификации"



conv2

### Сверточный слой (Convolution)

Двумерная свертка (2D convolution) - операция преобразования группы исходных пикселей (признаки) в меньшую группу пикселей (признаков).

Для этого используется фильтр (ядро-матрицу, окно свёртки) размером FxF (пример 3x3). Фильтр содержит веса для математического преобразования значений исходных пикселей в один итоговый.
- Исходная матрица 5x5 = 255 признаков
- Выходная матрица 3x3 = 9 признаков

conv1

Исходное изображение (W, H, Ch) имеет параметры:
- ширина в пикселях - W
- высота в пикселях - H
- количество цветовых каналов - Ch

Фильтр (ядро-матрица, окно свёртки) размером (F, F, Ch) (пример 3x3x1) имеет параметры:
- Количество каналов цвета - Сh
- Размер фильтра (ядра-матрицы, окна свёртки) (kernel_size) - F
- Значения ячеек фильтра (ядра-матрицы, окна свёртки, kernel)
- Фильтр (F, F, 1) имеет (F * F * 1) + 1 весов (добавляется свободный член)

Ниже представлена демонстрация работы фильтра размером (3, 3, 1) = 3x3x1 с весами

      [0, 1, 2]
      [2, 2, 0]
      [0, 1, 2]

conv2

Каждый сверточный (k@FxFxCh) слой `tf.keras.layers.Conv2D` имеет параметры:
- filters: количество фильтров - k
- kernel_size: размер, единый для всех фильтров слоя (ядр-матриц, окон свёртки, kernel) - F
- stride: размер шага окна свёртки
- padding: размер обрамления в ситуации, когда шаг и размером окна свёртки выходит за границы изображения
- значение весов свёрточного слоя = числовые значения в фильтрах.
    
Запись формата k@FxFxCh или k@FxFx (пример 6@5x5x1) означает,
    
    что свёрточный слой имеет k фильтров,
    каждый из которых - (F, F, Ch).
    
Количество весов вычисляется как (k * (F * F * Ch)) + (1 * k) весов (последнее "+1*k" -- так как у каждого фильтра есть ещё параметр смещения bias)

#### Padding

Padding добавляет дополнительные пиксели (обрамляет) для того, чтобы крайние пиксели исходгой матрицы могли попасть в цент ядра-матрицы.

conv2

#### Striding

Striding - пропуск/перескакивание некоторых областей исходного изображения. Достигается увеличением шага перемещения ядра-матрицы.

- Шаг stride=1 означает, что беруться пролеты через пиксель, то есть по факту каждый пролет является стандартной сверткой.
- Шаг stride=2 означает, что пролеты совершаются через каждые два пикселя, пропуская все другие пролеты в процессе и уменьшая их количество примерно в 2 раза
- Шаг stride=3 означает пропуск 3-х пикселей, сокращая количество в 3 раза и т.д.

conv2

### Объединяющий слой (Пулинг, Pooling)

Объединяющий слой (также употребляются названия Subsampling, Cубдискритизирующий, Пулинг, Pooling), схож со светрочным, но __фильтр (ядро-матрица, окно свёртки, kernel) этого слоя не имеет весов__, а использует агрегатную функцию. Применяется для уменьшения размера исходных изображений (признаков), уменьшения шумов, выделение доминирующего признака.

Выделяют два типа агрегатных функций:
- выбор максимального значения (Max Pooling)   из части изображения, покрываемой фильтром.
- вычисление среднего значения (Average Pooling) из части изображения, покрываемой фильтром.

Каждый объединяющий слой, `tf.keras.layers.AveragePooling2D` и `tf.keras.layers.MaxPooling2D`, имеет параметры:
- pool_size: размер окна фильтра
- strides: шаг окна фильтра
- padding: тип обрамления

conv2

Размер выходного изображения после свёртки или объединения вычисляется по следующей формуле
    
    ((imgSize - filterSize + 2*padding) // stride) +1,

где:
  - imgSize: размер высоты или ширины исходного изображения, пикс.
  - filterSize: размер окна, пикс.
  - padding: размер обрамления, пикс.
  - stride:  размер шага окна свёртки

# Создание свёрточной НС. Создание структуры и выбор начальных параметров

Создадим структуру свёрточной нейросети для которой определим:
- размер входного слоя `Input` равный разрешению изображения и кол-ву цветовых каналов (heihgt, width, colorChannels)
- одну или несколько пар слоёв:
    - сверточный `Conv2D` с указанием количества и размера окон свёртки
    - объединяющий `MaxPooling2D` либо `AveragePooling2D` с указанием размера объединяющее окна

- особый слой `Flatten`, задача которого развернуть двумерный массив изображения в одномерный вектор
- группу кол-ва скрытых полно связных слоёв `Dense` с нелинейными функциями активации
- выходной слой с определённым кол-во нейронов, равным кол-ву различаемых классов, и  с функцией активации `softmax` для преобразования ответов всех выходных нейронов НС в вероятности принадлежности к классам, так, чтобы сумма всех вероятностей = 1.0

Параметры [свёрточных слоёв](https://www.tensorflow.org/api_docs/python/tf/keras/layers) рассмотрены выше.

* Некоторые [функции активации](https://www.tensorflow.org/api_docs/python/tf/keras/activations) для скрытых слоев из пакета `tf.keras.activations`:
    - None
    - linear
    - relu
    - elu
    - sigmoid
    - tanh

# Простая модель

```python
model = tf.keras.Sequential()
model.add(layers.Input(shape=(28,28,1))) # shape=(height, width, colorChannels)

#========================== Блок выделения признаков ======================
model.add(layers.Conv2D(
         filters= 32,
         padding='same',
         kernel_size=(3,3),
         strides=1,
         activation=activations.relu
))
model.add(layers.MaxPooling2D(
        pool_size=(2,2),
        strides=2
))

#========================== Блок классификации ============================
model.add(layers.Flatten())
model.add(layers.Dense(units=128,activation=activations.relu))
model.add(layers.Dropout(0.5))

#========================== Выходной слой вычисления y_i ======================
model.add(layers.Dense(units=n_classes, activation=activations.softmax))
```

* Функцию оценки потерь (способ оценки ошибок между y_pred и y).
    - Для задачи многоклассовой классификации выберем `CategoricalCrossentropy`
    Доступные функиции оценки потерь смотрите на странице справки https://www.tensorflow.org/api_docs/python/tf/keras/losses


* Функцию оптимизации
    -  Выберем градиентный метод `Adam` или `SGD` (Стохастический градиентный спуск)
    Доступные функции оптимизации смотрите на странице справки https://www.tensorflow.org/api_docs/python/tf/keras/optimizers

```python
def compileModel(model):
    fLoss      = losses.CategoricalCrossentropy()
    fOptimizer = optimizers.Adam(learning_rate=0.0001)
    fMetric    = [metrics.CategoricalAccuracy()]
    model.compile(
        loss      = fLoss,
        optimizer = fOptimizer,
        metrics   = fMetric
    )
```

### Обучение НС

Для обучения используем функцию `model.fit()`, в процессе обучения будут изменяться весовые коэфф. сети так, чтобы входные значения `train_images` приводили к вычислению соответствующей ожидаемой метрики из `train_labels`. Функция `.fit()` имеет ряд дополнительных параметров (подробнее см. https://keras.io/api/models/model_training_apis/#fit-method )

*Примечание: Рекомендуется для начала установить кол-во эпох обучения `epochs` от 5 до 20. При необходимости вы можете запускать ячейку с функцией `model.fit()` повторно*

```python
def trainModelEpocs(model, epochs, totalHistoryLossTrain, totalHistoryLossTest):
    startFitTime  =  time.time()
    history = model.fit(
        x=train_images,
        y=train_labels,
        epochs=epochs,
        batch_size=32,
        verbose=1,
        validation_data=(test_images, test_labels)
    )
    deltaFitTime = time.time() - startFitTime

    totalHistoryLossTrain.extend(history.history['loss'])
    if 'val_loss' in history.history.keys():
        totalHistoryLossTest.extend(history.history['val_loss'])
    return deltaFitTime
```

```python
def trainModelFull(model, epochs, step):
    totalHistoryLossTrain = []
    totalHistoryLossTest  = []
    fullFitTime = 0
    while (epochs > 0):
        fullFitTime += trainModelEpocs(model, step, totalHistoryLossTrain, totalHistoryLossTest)
        epochs-=step
    return totalHistoryLossTrain, totalHistoryLossTest, fullFitTime
```

В задаче классификации объекты, обладающие искомым свойством, называют положительными (Positive), соответственно, объекты у которых искомое свойство отсутствует – отрицательными (Negative).

При совпадении вычисленного заключения модели `y_pred` с фактическим состоянием `y_true`, такое заключение получает статус истинного (True), при не совпадении - получает статус ложного заключения (False). В зависимости от комбинаций, возможны следующие ситуации-исходы:

    - истинно положительный (TP, True  Positive) - ВЕРНОЕ    вычисленное заключение о принадлежности к классу С ИСКОМЫМ свойством;  
    - ложно   положительный (FP, False Positive) - ОШИБОЧНОЕ вычисленное заключение о принадлежности к классу С ИСКОМЫМ свойством;
    - ложно   отрицательный (FN, False Negative) - ОШИБОЧНОЕ вычисленное заключение о принадлежности к классу БЕЗ ИСКОМОГО свойства;
    - истинно отрицательный (TN, True  Negative) - ВЕРНОЕ    вычисленное заключение о принадлежности к классу БЕЗ ИСКОМОГО свойства;

Одним из способов оценки точности моделей в задачах классификации является построение в графическом виде матрицы соответствия вычисленных ответов НС `y_pred_all_labels` ожидаемым истинным `y_test_all_labels`, для краткости -  построение **матрицы ошибок (confusion matrix)**. Матрица размером N*N, где N - кол-во различаемых классов. По осям матрицы находятся возможные метки классов. На горизонтальной оси - метки классов вычисленные моделью. На вертикальной оси - ожидаемые метки классов с правильными ответами. В каждой ячейке указано кол-во образцов с соответствующей комбинацией ответов. Идеальным вариантом является заполнение диагонали матрицы.

При наличии классов более 2 (многоклассовой классификации) матрицу оценивают относительно одного конкретного класса.

Рассмотрим пример построения матрицы ошибок для классификации, между тремя классами А,Б,В общим количеством по 10 образцов каждого, относительно класса Б.

# Построение эксперментов

```python
def report(model, epochs, step, wrongNum):
    totalHistoryLossTrain, totalHistoryLossTest, fullFitTime=trainModelFull(model, epochs, step)
    lossesPlot(totalHistoryLossTrain, totalHistoryLossTest)
    
    print('Вычисление ответов - вызов model.predict(X)')
    y_pred_all_probability = model.predict(test_images)
    y_pred_all_labels = tf.argmax(y_pred_all_probability, axis=1)
    y_test_all_labels = tf.argmax(test_labels, axis=1)
    calcAndPlotConfusionMatrix (
        y_test_all_labels,
        y_pred_all_labels,
        class_names
    )
    
    print(
    classification_report(
        y_true= y_test_all_labels,
        y_pred= y_pred_all_labels,
        target_names=class_names,
    ))
    
    mask = np.not_equal(y_test_all_labels.numpy(), y_pred_all_labels.numpy())
    indexesAll = np.arange(y_test_all_labels.shape[0])
    indexesWrongPredict = indexesAll[mask]
    for i, imgNum in enumerate(indexesWrongPredict):
        yiTrue = y_test_all_labels.numpy()[imgNum]
        yiPred = y_pred_all_labels.numpy()[imgNum]
        checkPrediction(model, test_images, test_labels, imgNum)
        if i>=wrongNum: break
    
```

```python
def trainForReport(model, epochs, step):
    steps_epochs = []
    remaining = epochs
    while remaining > 0:
        current_step = min(step, remaining)
        steps_epochs.append(current_step)
        remaining -= current_step

    # Хранилища данных
    all_loss = {'loss': [], 'val_loss': []}
    confusion_matrices = []
    precisions_over_time = []
    recalls_over_time = [] 
    epoch_marks = []

    current_epoch = 0

    for i, epochs_to_train in enumerate(steps_epochs):
        print(f"\n=== Шаг {i+1}: обучение {epochs_to_train} эпох (до эпохи {current_epoch + epochs_to_train}) ===")
        history = model.fit(
            train_images, train_labels,
            epochs=current_epoch + epochs_to_train,
            initial_epoch=current_epoch,
            validation_data=(test_images, test_labels),
            verbose=1
        )
        for key in all_loss:
            all_loss[key].extend(history.history[key])
        current_epoch += epochs_to_train
        epoch_marks.append(current_epoch)

        y_pred_prob = model.predict(test_images)
        y_pred = tf.argmax(y_pred_prob, axis=1).numpy()
        y_true = tf.argmax(test_labels, axis=1).numpy()

        cm = confusion_matrix(y_true, y_pred)
        confusion_matrices.append(cm)

        report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
        precisions_over_time.append([report[cls]['precision'] for cls in class_names])
        recalls_over_time.append([report[cls]['recall'] for cls in class_names])
        
    return all_loss, confusion_matrices, precisions_over_time, recalls_over_time, epoch_marks, y_true, y_pred
```

```python
def showReport(all_loss, confusion_matrices, precisions_over_time, recalls_over_time, epoch_marks, y_true, y_pred, wrongImgs=5):
    lossesPlot(all_loss['loss'], all_loss['val_loss'])
    num_steps = len(confusion_matrices)
    cols = min(3, num_steps)
    rows = (num_steps + cols - 1) // cols
    fig_cm, axes = plt.subplots(rows, cols, figsize=(cols*4, rows*4))
    if num_steps == 1:
        axes = [axes]
    else:
        axes = axes.flatten()

    for idx, cm in enumerate(confusion_matrices):
        ax = axes[idx]
        im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
        ax.set_title(f'After {epoch_marks[idx]} epochs')
        ax.set_xlabel('Predicted')
        ax.set_ylabel('True')
        # Визуализация чисел в ячейках
        thresh = cm.max() / 2.
        for i_row in range(cm.shape[0]):
            for j_col in range(cm.shape[1]):
                ax.text(j_col, i_row, format(cm[i_row, j_col], 'd'),
                        ha="center", va="center",
                        color="white" if cm[i_row, j_col] > thresh else "black")
    # Скрыть лишние подграфики
    for j in range(idx+1, len(axes)):
        fig_cm.delaxes(axes[j])
    fig_cm.tight_layout()
    plt.show()

    # Динамика precision и recall
    plt.figure(figsize=(12, 5))
    # Precision
    plt.subplot(1, 2, 1)
    for cls_idx, cls_name in enumerate(class_names):
        prec_vals = [prec[cls_idx] for prec in precisions_over_time]
        plt.plot(epoch_marks, prec_vals, marker='o', label=cls_name)
    plt.xlabel('Epoch')
    plt.ylabel('Precision')
    plt.title('Precision per class')
    plt.legend()
    plt.grid(True)

    # Recall
    plt.subplot(1, 2, 2)
    for cls_idx, cls_name in enumerate(class_names):
        rec_vals = [rec[cls_idx] for rec in recalls_over_time]
        plt.plot(epoch_marks, rec_vals, marker='o', label=cls_name)
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.title('Recall per class')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    print("\nФинальный отчет:")
    print(classification_report(y_true, y_pred, target_names=class_names))

    if wrongImgs > 0:
        print("\nПримеры неверных классификаций:")
        mask = y_true != y_pred
        wrong_indices = np.where(mask)[0]
        for idx in wrong_indices[:wrongImgs]:
            checkPrediction(model, test_images, test_labels, idx)
```

```python
def reportFull(model, epochs, step, wrongImgs=5):
    curTime = time.time()
    all_loss, confusion_matrices, precisions_over_time, recalls_over_time, epochs, y_true, y_pred = trainForReport(model, epochs, step)
    fullTime = time.time() - curTime
    showReport(all_loss, confusion_matrices, precisions_over_time, recalls_over_time, epochs, y_true, y_pred, wrongImgs)
    return fullTime
    
```

Точность сверточных сетей больше, чем у многослойного перцептрона, разница во времени зависит и от разницы в числе эпох и количестве параметров. Однако перцептрон начинает переобучаться уже к 40 эпохе, тогда как сверточные сети стабильно улучшают свою точность по обеим выборкам. Что приме-чательно, для таких сетей свойственна меньшая ошибка для тестовой вы-борки, чем у тренировочной. Большее количество блоков выделения признаков улучшает точность при том же классификаторе.

Слой Dropout для обоих типов сетей служит защитой от запоминания тренировочной части. Однако для перцептрона это не улучшает метрики, лишь замедляет обучение
